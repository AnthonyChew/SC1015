```python
pos, neg, total = 0, 0, 0
for i in data['positive']:
    if i == 0:
        pos += 1
        
for j in data['negative']:
    if j == 0:
        neg += 1



print(f"Positive: {pos}, Negative: {neg}")
```

```python
# review = pd.DataFrame(columns=['review'])
for i in range(41308):
    if data['positive'][i] == 0 and data['negative'][i] == 0:
        data['review'][i] = 0
    else:
        data['review'][i] = int((data['positive'][i] / (data['positive'][i] + data['negative'][i])) * 100)
    print(f"appid: {data['appid'][i]}, data['review'][i] : {data['review'][i]}, owners: {data['owners'][i]}")
    print(f"i: {i}\n")
```
