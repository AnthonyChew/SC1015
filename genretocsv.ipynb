```python
def duplicator(lst):
    check = []
    for i in lst:
        if i in check:
#             print("found duplicate! == ", i)
            continue
        else:
#             print("appending ", i)
            check.append(i)
    length = len(check)
    while (length < 7):
#         print("check length: ", length, " ----- check: ", check)
        check.append("NULL")
        length += 1
    return check

def countgenre(lst, counter):
    for i in lst:
        if i in counter:
            continue
        else:
            counter.append(i)
    return counter
```
