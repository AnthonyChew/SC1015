```python
def duplicator(lst):
    check = []
    for i in lst:
        if i in check:
#             print("found duplicate! == ", i)
            continue
        else:
#             print("appending ", i)
            check.append(i)
    length = len(check)
    while (length < 7):
#         print("check length: ", length, " ----- check: ", check)
        check.append("")
        length += 1
    return check

def countgenre(lst, counter):
    for i in lst:
        if i in counter:
            continue
        else:
            counter.append(i)
    return counter
```

```python
newgenre = pd.DataFrame(columns=['genre1', 'genre2', 'genre3', 'genre4', 'genre5', 'genre6', 'genre7'])
genrelist = []
counter = 1
# max = 0
for i in range(len(genre)):
    if genre['genre'][i] == '':
        strlist = []
    else:
        string = genre['genre'][i]
        strlist = string.split(', ')
#         genrelist = countgenre(strlist, genrelist)
    strlist = duplicator(strlist)
#     if len(strlist) > max:
#         max = len(strlist)
    strseries = pd.Series(strlist, index=newgenre.columns)
    newgenre = newgenre.append(strseries, ignore_index=True)
#     print(len(strlist), strlist)
#         print(strlist)
# print("max: ", max, "\n\n")
print(newgenre)
# print(genrelist)
```

```python
def languagefilter(lst, check):
    newlist = []
    print('\nList: ', lst, '\n')
    length = len(lst)
    for i in range(length):
        if "English" in lst[i]:
            check[0] = "Yes"
            lst[i] = "removed"
        elif "Chinese" in lst[i]:
            check[1] = "Yes"
            lst[i] = "removed"
        elif "French" in lst[i]:
            check[2] = "Yes"
            lst[i] = "removed"
        elif "German" in lst[i]:
            check[3] = "Yes"
            lst[i] = "removed"
        elif "Italian" in lst[i]:
            check[4] = "Yes"
            lst[i] = "removed"
        elif "Spanish" in lst[i]:
            check[5] = "Yes"
            lst[i] = "removed"
        
    for j in lst:
        print("j: ", j)
        if j == "removed":
            continue
        else:
            newlist.append(j)
    
    
    if check[5] == "Yes":
        newlist.insert(0, "Spanish")
    
    if check[4] == "Yes":
        newlist.insert(0, "Italian")
    
    if check[3] == "Yes":
        newlist.insert(0, "German")
    
    if check[2] == "Yes":
        newlist.insert(0, "French")
   
    if check[1] == "Yes":
        newlist.insert(0, "Chinese")
    
    if check[0] == "Yes":
        newlist.insert(0, "English")
    
    check[6] = len(newlist)
    check[7] = newlist
    print(check, '\n\n')
    return check
```
