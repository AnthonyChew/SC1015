Experimenting with different columns of the data to see which columns affect response variable owner_cat, AKA EDA of data

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics

# For Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('GameData_backup_with_review.csv' ,low_memory=0, header=0 , keep_default_na=False , encoding='ANSI')
# data.info()
```

```python
y = pd.DataFrame(data["owner_cat"])
x = pd.DataFrame(data[['positive', 'negative', 'average_forever', 'average_2weeks', 
                       'median_forever', 'median_2weeks', 
                       'initialprice', 'total_lang', '18+', 'Casual', 'Indie', 'RPG',
                       'Strategy', 'Sports', 'Simulation', 'Racing', 'Act_Adv',
                       'Massively Multiplayer', 'Free to Play', 'review']])
```

```python
f = plt.figure(figsize=(12, 12))
sb.catplot(y = "owner_cat", data = y, kind = "count")
```

```python
f = plt.figure(figsize=(12, 12))
sb.histplot(data = y['owner_cat'])
```

```python
f = plt.figure(figsize=(12, 12))

sb.histplot(data = data, y="owner_cat", x="positive")
```

```python
# dataprice = data[data["owner_cat"] < 6]
dataWith0 = data[data["owner_cat"] == 0]
dataWith0 = dataWith0[dataWith0["initialprice"] < 100]
dataWith0 = dataWith0[dataWith0["positive"] < 25000]

dataWithoutno0 = data[data["owner_cat"] > 0]
dataWithoutno0 = dataWithoutno0[dataWithoutno0["initialprice"] < 100]
dataWithoutno0 = dataWithoutno0[dataWithoutno0["positive"] < 25000]
```

```python
f, axes = plt.subplots(2, 1, figsize=(12, 12))
sb.boxplot(data = dataWith0, y="owner_cat", x="initialprice", orient='h', ax = axes[0])
sb.boxplot(data = dataWithoutno0, y="owner_cat", x="initialprice", orient='h', ax = axes[1])
```

```python
f, axes = plt.subplots(2, 1, figsize=(12, 12))
sb.boxplot(data = dataWith0, y="owner_cat", x="positive", orient='h', ax = axes[0])
sb.boxplot(data = dataWithoutno0, y="owner_cat", x="positive", orient='h', ax = axes[1])
```

```python
f, axes = plt.subplots(2, 1, figsize=(12, 12))

sb.boxplot(data = dataWith0, y="owner_cat", x="total_lang", orient='h', ax = axes[0])
sb.boxplot(data = dataWithoutno0, y="owner_cat", x="total_lang", orient='h', ax = axes[1])
```

```python
f, axes = plt.subplots(2, 1, figsize=(12, 12))

sb.histplot(data = dataWith0, x = 'total_lang', y = 'owner_cat', ax = axes[0])
sb.histplot(data = dataWithoutno0, x = 'total_lang', y = 'owner_cat', ax = axes[1])
```
